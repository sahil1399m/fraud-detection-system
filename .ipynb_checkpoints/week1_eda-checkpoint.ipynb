# Fraud Detection System — Week 1: Exploratory Data Analysis
**Dataset:** Kaggle Credit Card Fraud Detection  
**Goal:** Understand the data, find patterns, visualise class imbalance

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set_theme(style="whitegrid")
print("✅ Libraries loaded successfully!")
```

```python
df = pd.read_csv('data/creditcard.csv')
print(f"Dataset shape: {df.shape}")
print(f"Rows: {df.shape[0]:,}  |  Columns: {df.shape[1]}")
df.head()
```

```python
import os
print(os.getcwd())
```

```python
print("=== Data Types ===")
print(df.dtypes)
print("\n=== Missing Values ===")
print(df.isnull().sum())
print("\n=== Basic Stats ===")
df.describe()
```

## Class Distribution (Fraud vs Legit)
The most important check — how imbalanced is our dataset?

```python
class_counts = df['Class'].value_counts()
class_pct = df['Class'].value_counts(normalize=True) * 100

print("=== Transaction Counts ===")
print(f"Legitimate (0): {class_counts[0]:,}  ({class_pct[0]:.3f}%)")
print(f"Fraudulent  (1): {class_counts[1]:,}  ({class_pct[1]:.3f}%)")

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].bar(['Legitimate', 'Fraudulent'], class_counts.values,
            color=['#1D9E75', '#D85A30'], edgecolor='white', width=0.5)
axes[0].set_title('Transaction Count by Class', fontweight='bold')
axes[0].set_ylabel('Count')
for i, v in enumerate(class_counts.values):
    axes[0].text(i, v + 1000, f'{v:,}', ha='center', fontsize=11)

axes[1].pie(class_counts.values, labels=['Legitimate', 'Fraudulent'],
            colors=['#1D9E75', '#D85A30'], autopct='%1.3f%%', startangle=90)
axes[1].set_title('Class Distribution (%)', fontweight='bold')

plt.tight_layout()
plt.savefig('data/class_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Plot saved!")
```

## Transaction Amount Analysis
How do fraud and legit transactions differ by amount?

```python
fraud = df[df['Class'] == 1]
legit = df[df['Class'] == 0]

print("=== Amount Statistics ===")
print(f"\nLegitimate transactions:")
print(f"  Mean:   ${legit['Amount'].mean():.2f}")
print(f"  Median: ${legit['Amount'].median():.2f}")
print(f"  Max:    ${legit['Amount'].max():.2f}")
print(f"\nFraudulent transactions:")
print(f"  Mean:   ${fraud['Amount'].mean():.2f}")
print(f"  Median: ${fraud['Amount'].median():.2f}")
print(f"  Max:    ${fraud['Amount'].max():.2f}")

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].hist(legit['Amount'], bins=50, color='#1D9E75', alpha=0.7, edgecolor='white')
axes[0].set_title('Legit Transaction Amounts', fontweight='bold')
axes[0].set_xlabel('Amount ($)')
axes[0].set_ylabel('Count')
axes[0].set_xlim(0, 1000)

axes[1].hist(fraud['Amount'], bins=50, color='#D85A30', alpha=0.7, edgecolor='white')
axes[1].set_title('Fraud Transaction Amounts', fontweight='bold')
axes[1].set_xlabel('Amount ($)')
axes[1].set_xlim(0, 1000)

plt.tight_layout()
plt.savefig('data/amount_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Plot saved!")
```

## Time Analysis
When do fraudulent transactions happen most often?

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 6))

axes[0].plot(legit['Time'] / 3600, alpha=0.3, color='#1D9E75', linewidth=0.5)
axes[0].set_title('Legit Transactions over Time', fontweight='bold')
axes[0].set_xlabel('Hours')
axes[0].set_ylabel('Transaction Index')

axes[1].scatter(fraud['Time'] / 3600, fraud['Amount'],
                color='#D85A30', alpha=0.5, s=10)
axes[1].set_title('Fraud Transactions — Time vs Amount', fontweight='bold')
axes[1].set_xlabel('Hours')
axes[1].set_ylabel('Amount ($)')

plt.tight_layout()
plt.savefig('data/time_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Plot saved!")
```

## Feature Correlation with Fraud
Which features are most strongly linked to fraud?

```python
plt.figure(figsize=(16, 6))
corr = df.corr()['Class'].drop('Class').sort_values()

colors = ['#D85A30' if x < 0 else '#1D9E75' for x in corr.values]
plt.bar(corr.index, corr.values, color=colors, edgecolor='white', linewidth=0.5)
plt.title('Feature Correlation with Fraud (Class=1)', fontweight='bold', fontsize=14)
plt.xticks(rotation=90, fontsize=9)
plt.axhline(0, color='black', linewidth=0.8)
plt.ylabel('Correlation')
plt.tight_layout()
plt.savefig('data/feature_correlations.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n✅ Top 5 features POSITIVELY correlated with fraud:")
print(corr.tail(5).to_string())
print("\n✅ Top 5 features NEGATIVELY correlated with fraud:")
print(corr.head(5).to_string())
```

## Week 1 Findings Summary

| Finding | Value |
|---|---|
| Total transactions | 284,807 |
| Fraudulent transactions | 492 (0.172%) |
| Legitimate transactions | 284,315 (99.828%) |
| Max fraud amount | ~$2,125 |
| Mean fraud amount | ~$122 |
| Most correlated with fraud | V11, V4 (positive) |
| Most negatively correlated | V14, V12 (negative) |

### Key Takeaway
The dataset is **severely imbalanced** — only 0.172% fraud.
A naive model predicting "all legit" would be 99.8% accurate but completely useless.
We MUST handle this imbalance in Week 2 using SMOTE.
